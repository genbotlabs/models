```python
import re
import os

def preview_bracket_and_speaker_tags(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()
    
    # 1. ë³‘ê¸° í‘œí˜„ íŒ¨í„´
    bracket_pattern = re.compile(r'\(([^()]+?)\)/\(([^()]+?)\)')
    bracket_matches = list(bracket_pattern.finditer(text))

    # 2. í™”ì íƒœê·¸ íŒ¨í„´
    #   - ë„ì–´ì“°ê¸° ìˆëŠ” ê²½ìš°: ' o/'  or '\no/'
    #   - ë¶™ì–´ ìˆëŠ” ê²½ìš°: 'í¬ë§n/', 'ë§ì”€b/' ë“±
    tag_pattern = re.compile(r'\b[a-zA-Z]/|[ê°€-í£a-zA-Z]+[a-zA-Z]/')
    tag_matches = list(tag_pattern.finditer(text))

    if bracket_matches or tag_matches:
        print(f"\nğŸ“„ {os.path.basename(file_path)}:")

        if bracket_matches:
            print("  ğŸ”¹ ë³‘ê¸° í‘œí˜„ íƒì§€:")
            for match in bracket_matches:
                original = match.group(0)
                replacement = match.group(1).strip()
                print(f"   - ì›ë¬¸: {original} â†’ ì •ì œ ê²°ê³¼: {replacement}")

        if tag_matches:
            print("  ğŸ”¹ í™”ì íƒœê·¸ íƒì§€:")
            for match in tag_matches:
                context = match.group(0)
                print(f"   - ë°œê²¬ëœ íƒœê·¸: {context} (pos {match.start()})")

# â–¶ ìƒìœ„ 10ê°œ íŒŒì¼ë§Œ í…ŒìŠ¤íŠ¸ (í•„ìš”ì‹œ ìˆ˜ ì¡°ì ˆ)
for txt_path in labeled_data_list[:15]:
    preview_bracket_and_speaker_tags(txt_path)
```

```python
# ë³‘ê¸° í‘œí˜„ê³¼ ì´ìŠ¤ì¼€ì´í”„ë¬¸ ì œê±°
import re

def clean_text_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()

    # 1. ë³‘ê¸° í‘œí˜„ ì œê±°: (A)/(B) â†’ A
    text = re.sub(r'\(([^()]+?)\)/\(([^()]+?)\)', lambda m: m.group(1).strip(), text)

    # 2. í™”ì íƒœê·¸ ì œê±°: o/, n/, b/ ë“± (ë„ì–´ì“°ê¸° ìœ ë¬´ ê´€ê³„ì—†ì´ ì œê±°)
    #    ì˜ˆ: 'í¬ë§n/' â†’ 'í¬ë§', 'b/' â†’ '', ' n/' â†’ ' '
    text = re.sub(r'\b[a-zA-Z]/', '', text)  # ë…ë¦½ëœ íƒœê·¸ ì œê±° (ì˜ˆ: "b/")
    text = re.sub(r'([ê°€-í£a-zA-Z])([a-zA-Z])/', r'\1', text)  # ë‹¨ì–´ì— ë¶™ì€ íƒœê·¸ ì œê±° (ì˜ˆ: í¬ë§n/ â†’ í¬ë§)

    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(text)

# ëª¨ë“  í…ìŠ¤íŠ¸ íŒŒì¼ ì •ì œ ì ìš©
for txt_path in labeled_data_list:
    clean_text_file(txt_path)

print("âœ… ëª¨ë“  í…ìŠ¤íŠ¸ íŒŒì¼ ì •ì œ ì™„ë£Œ")
```
